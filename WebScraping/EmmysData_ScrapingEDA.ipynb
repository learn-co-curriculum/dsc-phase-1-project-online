```python
all_nominated_shows = nominated_shows + shows2019 + shows2018 + shows2017
```

#### Create Network/ Series DF
hoping that shows without a network (shows where only actors are nominated, not the show itself) can be filled in from other years

```python
#check to be sure lists are same length before creating a dictionary
print(len(all_nominated_shows))
print(len(networks))
```

```python
#create a dictionary
network_shows = dict(zip(all_nominated_shows, networks))

#dataframe from dictionary
network_df = pd.DataFrame.from_dict(network_shows,'index')
network_df.columns = ['Network']
network_df.head()
```

### Combine 2017-2020 DFs

```python
all_dfs = [joined_df, joined_df_2019, joined_df_2018, joined_df_2017]
combined_df = pd.concat(all_dfs)

#replace na with 0
combined_df.fillna('0', inplace=True)

#change to correct dtype
combined_df['Nominations']=combined_df['Nominations'].astype('int64')
combined_df['Actor Nominations']=combined_df['Actor Nominations'].astype('int64')

combined_df.head()
```

```python
#find total nominations by summing actor and show
combined_df['Total Nominations'] = combined_df['Nominations'] + combined_df['Actor Nominations']
combined_df.head()
```

### Add Network to the joined DFs

```python
awards_df = combined_df.join(network_df, how = 'left')
#drop any rows with no network info
awards_df['Network'].dropna(inplace=True)
#reorder columns
awards_df = awards_df[['Network', 'Award Year','Nominations', 
                       'Actor Nominations', 'Total Nominations']]
#awards_df.astype({'Nominations':'int64', 'Actor Nominations':'int64'})
awards_df
```
